### Quick and dirty: In this notebook, we look at who wrote the articles and whether any of the named key actors wrote any

```python
import os
import pandas as pd

DIR_DATA_INT = os.environ.get("DIR_DATA_INTERIM")
b1 = pd.read_csv(os.path.join(DIR_DATA_INT, "uk_news.csv"))
b2 = pd.read_csv(os.path.join(DIR_DATA_INT, "uk_news_sample2.csv"))
```

```python
key_actors = ['chater', 'halpern', 'michie', 'sunstein', 'kahneman', 'thaler']
```

### Batch 1

List of all unique authors

```python
b1_authors = b1.byline.unique()
```

```python
b1_authors_list = [t.lower() for t in b1_authors if isinstance(t, str)]
```

```python
# let's see them
b1_authors_list
```

```python
[[authors for authors in b1_authors_list if actor in authors] for actor in key_actors]
```

Only Nick Chater authored one article in batch 1.

```python
# let get the metadata about the article authored by N.Chater
b1[b1.byline.isin(['Nick Chater'])]
```

### Batch 2

List of all unique authors

```python
b2_authors = b2.byline.unique()
```

```python
b2_authors_list = [t.lower() for t in b2_authors if isinstance(t, str)]
```

```python
# let's see them
b2_authors_list
```

```python
[[authors for authors in b2_authors_list if actor in authors] for actor in key_actors]
```

No key actor authored any article in batch 2.
